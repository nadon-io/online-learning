## Python syntax

This guide is written with the assumption that the reader has some experience in scientific computer or computer programming. Here are some key features in Python to keep in mind:

1. Blocks (such as for-loops, if-statements, function definitions) are started with the expression followed by a colon, with the interior indented, and no end statement

Example:

```Python
if 5 > 2:
    print("this is true")
```

2. Indexing is 0-based, C-style. This means the first element of `my_list` is accessed at `my_list[0]`
3. Everything is an object. Contents of objects are accessed via the `.` syntax. For example: `my_object.val`. or `my_object.my_func(3)`
4. Python has "duck-typing", which means you don't usually specify the types of your variables and instead just let the interpreter figure out what the types should be. For example: `2 / 3` in Python results in the float `0.33`..., while in C this would result in the integer `0`.
5. While things like ranges in languages such as Matlab or Julia are (inclusive, inclusive), Python is (inclusive, exclusive). This means `my_arr[0, 10]` will fetch the first 10 elements (0, 1, ..., 9) and *exclude* `my_arr[10]`.

## Using packages

Let's do an example within a Jupyter Notebook. Since all the cells are still connected to eachother, if the cell above with the imports was ran we should be able to use the packages below:

````python
arr = np.arange(1, 101, 2) # numbers from 1 to 101 (exclusive), by steps of 2
arr
````

````python
# If you want to pass an array into a function, make sure to use a "vectorized" function
arr2 = np.log(arr)
arr2
````

````python
# my_data is a Python Dictionary, which is one of the more powerful Python features.
# Python Dictionaries are very similar to JSON Objects.
my_data = {"x": arr, "my_exp": arr2}

# Below is how you create a Pandas DataFrame
# Usually you just pass in the dictionary directly like below.
# Of course the following would still work: pd.DataFrame(my_data)
df = pd.DataFrame({"x": arr, "my_exp": arr2})
df
````

We just saw how we could use Numpy to work with arrays, matrices, and other numerical features. Although Python's `math` module and numpy both share similar functions (eg. `exp`), you should always use Numpy's version as it offers better compatibility with other Numpy objects such as arrays.

We also saw how we could transform our data into a DataFrame using Pandas. I cannot stress how powerful Pandas is, as it includes pretty much everything you could want for working with data (including plotting, applying functions, filtering, loading/saving...). See the documentation [here](https://pandas.pydata.org/pandas-docs/stable/).

However in terms of plotting Matplotlib is **the** golden standard among the scientific community, to the point where other languages such as Julia offer access to it via an internal call to Python code. Matplotlib provides the most robust, extensive set of features for plotting and so when doing important work such as publishing a paper it should always be preferred over others (even Matlab's own commercial solution!).

Seaborn is essentially a simplified wrapper for Matplotlib, and thus gives us access to great features while also making it easier.

````python
# The figure (used so we can save it later)
fig = plt.figure()

# The colorscheme to use, totally optional
cmap = sns.cubehelix_palette(rot=-.2, as_cmap=True)

# plot axes
ax = sns.scatterplot(x="x", y="my_exp",
                     hue="my_exp", size="my_exp",
                     palette=cmap, sizes=(200, 10),
                     data=df)
````
